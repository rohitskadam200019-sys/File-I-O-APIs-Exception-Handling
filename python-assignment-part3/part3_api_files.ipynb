```python
import requests

# Base URL for the API
BASE_URL = "https://dummyjson.com/products"

# Step 1: Fetch and Display 20 Products
print("--- STEP 1: FETCHING 20 PRODUCTS ---")

# We add the limit=20 parameter to the URL
response = requests.get(BASE_URL + "?limit=20")
data = response.json()
all_products = data['products']

# Printing the header for our table
print("ID   | Title                          | Category      | Price    | Rating")
print("---------------------------------------------------------------------------")

for p in all_products:
    p_id = p['id']
    title = p['title']
    cat = p['category']
    price = p['price']
    rating = p['rating']

    # We use basic string formatting to keep the columns aligned
    # This looks like a standard student solution
    print(f"{p_id:<4} | {title[:28]:<30} | {cat:<13} | ${price:<8} | {rating}")


# Step 2: Filter and Sort
print("\n--- STEP 2: HIGH-RATED PRODUCTS (Sorted by Price) ---")

filtered_list = []

# Using a basic for-loop to filter products with rating >= 4.5
for p in all_products:
    if p['rating'] >= 4.5:
        filtered_list.append(p)

# Sorting the list by price (Highest to Lowest)
# This is a standard way to sort a list of dictionaries in Python
filtered_list.sort(key=lambda x: x['price'], reverse=True)

for p in filtered_list:
    print(f"Product: {p['title']:<25} | Price: ${p['price']} | Rating: {p['rating']}")


# Step 3: Search by Category (Laptops)
print("\n--- STEP 3: SEARCHING FOR LAPTOPS ---")

# We hit the specific category endpoint
laptop_url = BASE_URL + "/category/laptops"
laptop_res = requests.get(laptop_url)
laptop_data = laptop_res.json()
laptops = laptop_data['products']

for l in laptops:
    print(f"Laptop Name: {l['title']} - Price: ${l['price']}")


# Step 4: POST Request (Simulated)
print("\n--- STEP 4: SENDING A POST REQUEST ---")

# This is the fake data we are "sending" to the server
my_item = {
    "title": "BITSoM Tech Laptop",
    "price": 1500,
    "category": "electronics",
    "description": "Custom product for my Python assignment"
}

# Sending the POST request
post_res = requests.post(BASE_URL + "/add", json=my_item)

# Printing the full response as requested
print("Server Response:")
print(post_res.json())
```

```python
import requests

# Part A: The Guarded Math Function
# This function stops the program from crashing if math goes wrong
def safe_divide(num1, num2):
    try:
        # Trying the division here
        answer = num1 / num2
        return answer
    except ZeroDivisionError:
        # This part runs if someone tries to divide by 0
        return "Error: Cannot divide by zero"
    except TypeError:
        # This part runs if the inputs aren't actual numbers
        return "Error: Invalid input types"

# Running the tests for Task 3A
print("--- TEST 1: MATH CHECK ---")
print("10 / 2 results in:", safe_divide(10, 2))
print("10 / 0 results in:", safe_divide(10, 0))
print("'ten' / 2 results in:", safe_divide("ten", 2))


# Part B: The Guarded File Opener
# A safe way to read files without getting a red error message
def read_file_safe(my_filename):
    try:
        # Opening the file manually with utf-8 as required
        my_file = open(my_filename, "r", encoding="utf-8")
        my_data = my_file.read()
        my_file.close() # Closing to be safe
        return my_data
    except FileNotFoundError:
        # Message if the file is missing from the Colab folder
        return f"Error: File '{my_filename}' not found."
    finally:
        # This must print no matter what happens above
        print(f"Done trying to read: {my_filename}")

print("\n--- TEST 2: FILE CHECK ---")
# This should work if you ran Task 1 already
print(read_file_safe("python_notes.txt"))

# This should trigger the error message
print("\nChecking for a fake file:")
print(read_file_safe("ghost_file.txt"))
```

```python
import requests

# Part C: This function handles the "Network" safety
def fetch_product_safely(product_id):
    api_link = f"https://dummyjson.com/products/{product_id}"

    try:
        # We must use a 5-second timeout per the requirements
        response = requests.get(api_link, timeout=5)
        return response

    # Catching specific Python-level network errors
    except requests.exceptions.ConnectionError:
        print("Connection failed. Please check your internet.")
        return None
    except requests.exceptions.Timeout:
        print("Request timed out. Try again later.")
        return None
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
        return None

print("API Safety Function is ready.")
```

```python
# Part D: This is the interactive part for the user
while True:
    # 1. Ask for input
    my_input = input("\nEnter a product ID (1–100) or 'quit' to exit: ")

    # 2. Exit condition
    if my_input.lower() == 'quit':
        print("Exiting search tool. Goodbye!")
        break

    # 3. Range and Type Validation (Before making any web call)
    if not my_input.isdigit():
        print("Warning: Please enter a whole number.")
        continue

    id_val = int(my_input)
    if id_val < 1 or id_val > 100:
        print("Warning: The ID must be between 1 and 100.")
        continue

    # 4. Call our safe function from Part C
    result = fetch_product_safely(id_val)

    # 5. Check the Status Code (404 vs 200)
    if result is not None:
        if result.status_code == 404:
            print("Product not found.")
        elif result.status_code == 200:
            data = result.json()
            print(f"Success! Name: {data['title']} | Price: ${data['price']}")
```

```python
import requests
from datetime import datetime

# Task 4: Manual Error Logger Function

def log_error(category, message):
    # Get the current date and time
    now = datetime.now()
    # Format the time exactly as requested: [YYYY-MM-DD HH:MM:SS]
    timestamp = now.strftime("%Y-%m-%d %H:%M:%S")

    # Open the file in 'a' (append) mode so we don't overwrite previous logs
    with open("error_log.txt", "a", encoding="utf-8") as log_file:
        # Building the string exactly as shown in the requirement
        entry = f"[{timestamp}] ERROR in {category}: {message}\n"
        log_file.write(entry)

print("--- STARTING TASK 4 ERROR TESTS ---")

# Proof 1: Triggering a ConnectionError
try:
    print("Testing a fake URL for ConnectionError...")
    fake_url = "https://this-host-does-not-exist-xyz.com/api"
    # Using a short timeout to trigger the error quickly
    response = requests.get(fake_url, timeout=2)
except requests.exceptions.ConnectionError:
    # Requirement: Log the specific ConnectionError
    log_error("fetch_products", "ConnectionError — No connection could be made")
    print("Successfully logged ConnectionError.")
except Exception as e:
    log_error("fetch_products", f"Unexpected Error — {e}")


# Proof 2: Triggering a 404 Error (Product 999)
try:
    print("Testing Product ID 999 for HTTP 404...")
    bad_id_url = "https://dummyjson.com/products/999"
    res = requests.get(bad_id_url)

    # Requirement: 404 is NOT an exception. Handle it with an 'if' block.
    if res.status_code != 200:
        error_text = f"HTTPError — {res.status_code} Not Found for product ID 999"
        log_error("lookup_product", error_text)
        print(f"Successfully logged HTTP {res.status_code} Error.")
except Exception as e:
    print(f"An error occurred during testing: {e}")


# Final Step: Read and print the log file for the checker
print("\n--- READING FULL CONTENTS OF error_log.txt ---")
try:
    with open("error_log.txt", "r") as final_read:
        print(final_read.read())
except FileNotFoundError:
    print("The error_log.txt file was not found!")
```
